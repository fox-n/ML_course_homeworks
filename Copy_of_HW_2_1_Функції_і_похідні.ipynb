1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = sp.symbols('x')
f_x1 = 10 * x - 5
der_f_x1 = sp.diff(f_x1, x)
print(f"Похідна лінійної функції 10 * x - 5:\n {der_f_x1}")
```

```python
def f(x):
  return 10 * x - 5
```

```python
xs = np.linspace(0,10,11)
df = np.full_like(xs, der_f_x1)
plt.plot(xs, f(xs), 'o', label='f(x) = 10 * x - 5', color='blue');
plt.plot(df, label="f'(x) = 10", color='green');

plt.title("Графік функції та її похідної")
plt.xlabel("Години на навчання")
plt.ylabel("Оцінка")
plt.legend()
plt.grid(True)

plt.show()
```

Висновок: Між годинами навчання та оцінкою є лінійне зростання зі сталою швидкістю. Проте зміна кількості годин навчання ніяк не пришвидшує зростання оцінки.

```python
f_x2 = sp.sin(x)
der_f_x2 = sp.diff(f_x2, x)
print(f"Похідна лінійної функції sin(x):\n {der_f_x2}")
```

```python
xs = np.linspace(0,10,100)
plt.plot(xs, np.log(xs), label='f(x) = ln(x)', color='blue');
plt.plot(xs, (1/(xs)), label="f'(x) = 1/x", color='green');

plt.title("Графік функції та її похідної")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
funct = x ** 2 - 5 * x + 1
der_funct = sp.diff(funct, x)
print(f"Похідна лінійної функції x ** 2 - 5 * x + 1:\n {der_funct}")
```

```python
def manuf_expenses(x):
  return x ** 2 - 5 * x + 1
```

```python
def manuf_expenses_der(x):
  return 2*x - 5
```

```python
xs = np.linspace(0,100,100)

plt.plot(xs, manuf_expenses(xs), label='f(x) = x ** 2 - 5 * x + 1', color='blue');
plt.plot(xs,manuf_expenses_der(xs), label="f'(x) = 2*x - 5", color='green');

plt.title("Витрати компанії vs Обсяги виробництва")
plt.xlabel("Кількість вироблених одиниць")
plt.ylabel("Загальні витрати компанії")
plt.legend()
plt.grid(True)

plt.show()
```
